```python
from pymongo import MongoClient
import pandas as pd

# --- Connection details ---
mongo_uri = "mongodb://172.19.19.87:27017/"
db_name = "MEDICATIONPAGE_100"
coll_name = "DATA2"

# --- Connect to Mongo ---
client = MongoClient(mongo_uri)
db = client[db_name]
collection = db[coll_name]

# --- Fetch all documents ---
docs = list(collection.find({}))   # {} = no filter, fetch all

# --- Convert to DataFrame ---
if docs:
    df = pd.DataFrame(docs)

    # Drop MongoDB internal _id if not needed
    if "_id" in df.columns:
        df = df.drop(columns=["_id"])

    # --- Save to Excel ---
    output_file = r".\All_Medication_Order\Medication_Data_ALL_Charts_DATA3.xlsx"
    df.to_excel(output_file, index=False)

    print(f"✅ Saved {len(df)} records to {output_file}")
else:
    print("⚠️ No data found in the collection.")
```
